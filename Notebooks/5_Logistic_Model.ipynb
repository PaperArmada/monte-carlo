### **6. Simulating Population Growth: The Logistic Model**

**Concept:**

Monte Carlo methods can simulate stochastic population dynamics, accounting for random events like birth and death rates.

**Simulation Steps:**

1. **Define the Logistic Growth Model:**

   - \( $\frac{dP}{dt} = rP\left(1 - \frac{P}{K}\right)$ \)
   
2. **Introduce Stochasticity:**

   - Add randomness to growth rate \( $r$ \) or carrying capacity \( $K$ \).
   
3. **Run Simulations:**

   - Iterate over time steps, updating population with random fluctuations.
   
4. **Analyze Outcomes:**

   - Examine how variability affects long-term population size.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monte_carlo_population_growth(P0, r_mean, r_std, K, t_max, dt, num_simulations):
    t = np.arange(0, t_max, dt)
    results = []
    for _ in range(num_simulations):
        P = np.zeros_like(t)
        P[0] = P0
        r = np.random.normal(r_mean, r_std, size=t.shape)
        for i in range(1, len(t)):
            P[i] = P[i-1] * (1 + r[i-1] * (1 - P[i-1]/K) * dt)
        results.append(P)
    return t, results
```

```python
# Parameters
P0 = 10         # Initial population
r_mean = 0.1    # Mean growth rate
r_std = 0.05    # Standard deviation of growth rate
K = 100         # Carrying capacity
t_max = 100     # Maximum time
dt = 0.1        # Time step
num_simulations = 100
```

```python
# Run simulation
t, results = monte_carlo_population_growth(P0, r_mean, r_std, K, t_max, dt, num_simulations)
```

```python
# Plot results
for P in results:
    plt.plot(t, P, color="gray", alpha=0.3)
plt.title('Stochastic Population Growth Simulations')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.show()
```

**Explanation:**

- **Stochastic Growth Rate:** Randomizes \( $r$ \) at each time step to simulate environmental variability.
- **Multiple Simulations:** Runs several iterations to observe different possible outcomes.
- **Visualization:** Shows the spread of population trajectories over time.

**Benefits:**

- Models real-world scenarios where population dynamics are influenced by random factors.
- Helps in assessing risks of extinction or overpopulation.
