```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_pi(num_samples):
    # Generate random points
    x = np.random.uniform(0, 1, num_samples)
    y = np.random.uniform(0, 1, num_samples)

    # Determine if points are inside the quarter circle
    inside_circle = x**2 + y**2 <= 1

    # Estimate Pi
    pi_estimate = 4 * np.sum(inside_circle) / num_samples
    return pi_estimate, x, y, inside_circle
```

```python
num_samples = 50_000
pi_estimate, x, y, inside_circle = estimate_pi(num_samples)
print(f"Estimated Pi: {pi_estimate}")
```

```python
# Visualization
plt.figure(figsize=(6,6))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1)
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1)
plt.title("Monte Carlo Simulation for Pi Estimation")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
