```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(num_steps):
    steps = np.random.choice([-1, 1], size=num_steps)
    position = np.cumsum(steps)
    return position
```

```python
num_steps = 1000
position = random_walk(num_steps)
```

```python
# Visualization
plt.plot(position)
plt.title('1D Random Walk')
plt.xlabel('Step')
plt.ylabel('Position')
plt.show()
```
